```python
def dataloader(dataPath):
    first = True
    images = []
    with open(dataPath+"Train.csv") as f:
        for line in f.readlines():
            if first:
                first=False
                continue
            _, _, _,_,_,_, classId, path = line.strip().split(",")
            images.append((classId, path))
    return images
```

```python
images = dataloader("./archive/")
```

```python
import pandas as pd
df = pd.DataFrame(images, columns=["class", "path"])
```
